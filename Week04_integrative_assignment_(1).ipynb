```python
# Import NumPy
import numpy as np

# Set random seed for reproducibility (so everyone gets the same data)
np.random.seed(4641)

# Generate simulated daily returns data
# 20 stocks, 252 trading days (1 year)
# Returns are in decimal form (e.g., 0.02 = 2% return)

stock_names = np.array([
    'APEX', 'BOLT', 'CYBER', 'DATA', 'EDGE',
    'FLUX', 'GRID', 'HYPER', 'INTEL', 'JADE',
    'KITE', 'LOGIC', 'MATRIX', 'NEXUS', 'ORBIT',
    'PULSE', 'QUANT', 'RELAY', 'SPARK', 'TITAN'
])

# Generate returns with different volatility profiles
# High-growth tech stocks have higher volatility
volatilities = np.array([0.02, 0.025, 0.03, 0.018, 0.022,
                         0.028, 0.015, 0.035, 0.02, 0.025,
                         0.032, 0.019, 0.024, 0.021, 0.027,
                         0.023, 0.031, 0.017, 0.026, 0.029])

# Expected annual returns (varying by stock)
expected_returns = np.array([0.12, 0.15, 0.18, 0.10, 0.14,
                             0.16, 0.08, 0.22, 0.11, 0.13,
                             0.20, 0.09, 0.15, 0.12, 0.17,
                             0.14, 0.19, 0.07, 0.16, 0.21])

# Generate 252 days of returns for each stock
# Shape: (252, 20) - rows are days, columns are stocks
daily_returns = np.random.normal(
    loc=expected_returns/252,  # Daily expected return
    scale=volatilities,        # Daily volatility
    size=(252, 20)
)

# Portfolio weights (how much is invested in each stock)
# These sum to 1.0 (100% of portfolio)
weights = np.array([0.08, 0.06, 0.07, 0.05, 0.04,
                    0.05, 0.04, 0.06, 0.05, 0.04,
                    0.05, 0.04, 0.06, 0.05, 0.05,
                    0.05, 0.06, 0.04, 0.03, 0.03])

print("Data loaded successfully!")
print(f"Daily returns shape: {daily_returns.shape}")
print(f"Number of stocks: {len(stock_names)}")
print(f"Trading days: {daily_returns.shape[0]}")
print(f"Portfolio weights sum: {weights.sum():.2f}")
```

## Part 1: Individual Stock Analysis (20 points)

Sarah needs to understand the risk profile of each stock in the portfolio.

**Your Task:**
Using NumPy array operations (not loops), calculate for each stock:
1. Average daily return
2. Daily return standard deviation (volatility)
3. Annualized return (average daily return × 252)
4. Annualized volatility (daily std × √252)
5. Sharpe Ratio: (Annualized Return - Risk-Free Rate) / Annualized Volatility
   - Use risk-free rate = 5% (0.05)

**Identify:**
- The stock with the highest annualized return
- The stock with the lowest volatility
- The stock with the best risk-adjusted return (highest Sharpe ratio)

```python
# Part 1: Individual Stock Analysis
# Use NumPy operations - NO LOOPS

risk_free_rate = 0.05

# Calculate statistics for each stock
# Hint: Use axis parameter to operate across the correct dimension

# Average daily returns (one value per stock)
avg_daily_return = np.mean(daily_returns, axis=0)

# Daily standard deviation
daily_volatility = np.std(daily_returns, axis=0)

# Annualized return
ann_return = avg_daily_return * 252

# Annualized volatility
ann_volatility = daily_volatility * np.sqrt(252)

# Sharpe Ratio
sharpe_ratio = (ann_return - risk_free_rate) / ann_volatility

# Display results
print("INDIVIDUAL STOCK ANALYSIS")
print("="*70)
print(f"{'Stock':<8} {'Ann. Return':>12} {'Ann. Vol':>12} {'Sharpe':>10}")
print("-"*70)

# Print each stock's metrics purely vectorized using np.column_stack
display_matrix = np.column_stack((
    stock_names,
    np.round(ann_return, 4).astype(str),
    np.round(ann_volatility, 4).astype(str),
    np.round(sharpe_ratio, 4).astype(str)
))
print('\n'.join(['\t\t'.join(row) for row in display_matrix]))

# Identify best performers
print("\n" + "="*70)
print("KEY FINDINGS")
print("-"*70)

best_return_idx = np.argmax(ann_return)
lowest_vol_idx = np.argmin(ann_volatility)
best_sharpe_idx = np.argmax(sharpe_ratio)

print(f"Highest Return Stock:      {stock_names[best_return_idx]} ({ann_return[best_return_idx]:.2%})")
print(f"Lowest Volatility Stock:   {stock_names[lowest_vol_idx]} ({ann_volatility[lowest_vol_idx]:.2%})")
print(f"Best Sharpe Ratio Stock:   {stock_names[best_sharpe_idx]} ({sharpe_ratio[best_sharpe_idx]:.2f})")
```

## Part 2: Portfolio Return Analysis (20 points)

Now Jordan needs to analyze the portfolio as a whole, considering how the weights affect overall performance.

**Your Task:**
1. Calculate the daily portfolio returns (weighted sum of individual returns)
   - Hint: For each day, multiply each stock's return by its weight and sum
2. Calculate the portfolio's:
   - Average daily return
   - Daily volatility
   - Annualized return
   - Annualized volatility
   - Sharpe ratio
3. Find the best and worst trading days
4. Calculate cumulative returns (how $1 invested would grow)

```python
# Part 2: Portfolio Return Analysis

# Calculate daily portfolio returns
# Hint: multiply daily_returns by weights (broadcasts across the 20 columns),
#       then sum across the stock axis: (daily_returns * weights).sum(axis=1)
portfolio_returns = (daily_returns * weights).sum(axis=1)

# Calculate portfolio statistics
port_avg_daily = np.mean(portfolio_returns)
port_daily_vol = np.std(portfolio_returns)
port_ann_return = port_avg_daily * 252
port_ann_vol = port_daily_vol * np.sqrt(252)
risk_free_rate = 0.05
port_sharpe = (port_ann_return - risk_free_rate) / port_ann_vol

# Find best and worst days
best_day_idx = np.argmax(portfolio_returns)
worst_day_idx = np.argmin(portfolio_returns)
best_day_return = portfolio_returns[best_day_idx]
worst_day_return = portfolio_returns[worst_day_idx]

# Calculate cumulative returns
# If you invested $1 on day 0, what would it be worth each day?
# Cumulative return: start a running value at 1.0 and multiply it by (1 + that day's
# portfolio return) one day at a time (a for-loop over the days is fine here).
cumulative_returns = np.zeros(len(portfolio_returns))
current_value = 1.0
for i in range(len(portfolio_returns)):
    current_value *= (1 + portfolio_returns[i])
    cumulative_returns[i] = current_value

# Display portfolio summary
print("PORTFOLIO PERFORMANCE SUMMARY")
print("="*60)
print(f"Annualized Return: {port_ann_return:.2%}")
print(f"Annualized Volatility: {port_ann_vol:.2%}")
print(f"Sharpe Ratio: {port_sharpe:.4f}")
print(f"\nBest Day: Day {best_day_idx} ({best_day_return:.2%})")
print(f"Worst Day: Day {worst_day_idx} ({worst_day_return:.2%})")
print(f"\nStarting Value: $1.00")
print(f"Ending Value: ${cumulative_returns[-1]:.2f}")
```

> I utilized NumPy broadcasting to multiply the 2D `daily_returns` array by the 1D `weights` array. This automatically applies each stock's weight to its respective column across all 252 days simultaneously, allowing me to sum across `axis=1` to get the daily portfolio returns without needing a nested loop. For the cumulative returns, I used a standard `for` loop as explicitly permitted, since calculating compounding returns iteratively depends on the prior day's value.

## Part 3: Risk Metrics (25 points)

The investment committee is particularly interested in downside risk—how bad could things get?

**Your Task:**
Calculate the following risk metrics:

1. **Value at Risk (VaR) at 95% confidence**
   - What is the 5th percentile of daily returns?
   - Interpretation: "On 95% of days, the portfolio won't lose more than X%"

2. **Conditional VaR (Expected Shortfall)**
   - Average of returns that are worse than the VaR
   - Interpretation: "When losses exceed VaR, the average loss is X%"

3. **Maximum Drawdown**
   - The largest peak-to-trough decline in cumulative value
   - Hint: Calculate running maximum of cumulative returns, then find biggest drop

4. **Downside Deviation**
   - Standard deviation of only the negative returns
   - More focused measure of "bad" volatility

5. **Sortino Ratio**
   - Like Sharpe, but uses downside deviation instead of total volatility
   - (Annualized Return - Risk-Free Rate) / Annualized Downside Deviation

```python
# Part 3: Risk Metrics

# 1. Value at Risk (95%)
# Hint: Use np.percentile()
var_95 = np.percentile(portfolio_returns, 5)

# 2. Conditional VaR (Expected Shortfall)
# Average of returns below VaR
mask_var = portfolio_returns < var_95
expected_shortfall = np.mean(portfolio_returns[mask_var])

# 3. Maximum Drawdown
# Calculate cumulative returns first, then the running maximum (track the largest
# cumulative value seen so far with a for-loop), then the drawdowns
running_max = np.zeros(len(cumulative_returns))
current_max = cumulative_returns[0]
for i in range(len(cumulative_returns)):
    if cumulative_returns[i] > current_max:
        current_max = cumulative_returns[i]
    running_max[i] = current_max

drawdowns = (cumulative_returns - running_max) / running_max
max_drawdown = np.min(drawdowns)

# 4. Downside Deviation
# Standard deviation of only negative returns
negative_returns = portfolio_returns[portfolio_returns < 0]
downside_dev = np.std(negative_returns)
ann_downside_dev = downside_dev * np.sqrt(252)

# 5. Sortino Ratio
sortino_ratio = (port_ann_return - risk_free_rate) / ann_downside_dev

# Display risk report
print("RISK ANALYSIS REPORT")
print("="*60)
print("\nDownside Risk Metrics:")
print("-"*60)
print(f"Value at Risk (95%): {var_95:.2%}")
print(f"  → On 95% of days, daily loss won't exceed this amount")
print(f"\nExpected Shortfall: {expected_shortfall:.2%}")
print(f"  → When VaR is breached, average loss is this amount")
print(f"\nMaximum Drawdown: {max_drawdown:.2%}")
print(f"  → Largest peak-to-trough decline over the period")
print(f"\nDownside Deviation (annualized): {ann_downside_dev:.2%}")
print(f"\nSortino Ratio: {sortino_ratio:.4f}")
print(f"  → Risk-adjusted return using only downside volatility")
```

## Part 4: Weight Optimization Scenario (20 points)

Sarah asks Jordan to explore what would happen with different portfolio weights.

**Scenario: Equal Weight Portfolio**
What if instead of the current weights, we invested equally in all 20 stocks (5% each)?

**Your Task:**
1. Create equal weights (5% each stock)
2. Calculate the equal-weight portfolio's daily returns
3. Compare the equal-weight vs current portfolio:
   - Annualized return
   - Annualized volatility
   - Sharpe ratio
   - Maximum drawdown
4. Identify which stocks are overweight and underweight in the current portfolio compared to equal weight
5. Provide a brief interpretation: Which portfolio is better and why?

```python
# Part 4: Weight Optimization Scenario

# Create equal weights
num_stocks = len(stock_names)
eq_weights = np.full(num_stocks, 1.0 / num_stocks)

# Calculate equal-weight portfolio returns
eq_portfolio_returns = (daily_returns * eq_weights).sum(axis=1)

# Calculate metrics for equal-weight portfolio
eq_ann_return = np.mean(eq_portfolio_returns) * 252
eq_ann_vol = np.std(eq_portfolio_returns) * np.sqrt(252)
eq_sharpe = (eq_ann_return - risk_free_rate) / eq_ann_vol

# Calculate equal-weight max drawdown
eq_cum_returns = np.zeros(len(eq_portfolio_returns))
eq_current_val = 1.0
for i in range(len(eq_portfolio_returns)):
    eq_current_val *= (1 + eq_portfolio_returns[i])
    eq_cum_returns[i] = eq_current_val

eq_running_max = np.zeros(len(eq_cum_returns))
eq_current_max = eq_cum_returns[0]
for i in range(len(eq_cum_returns)):
    if eq_cum_returns[i] > eq_current_max:
        eq_current_max = eq_cum_returns[i]
    eq_running_max[i] = eq_current_max

eq_drawdowns = (eq_cum_returns - eq_running_max) / eq_running_max
eq_max_drawdown = np.min(eq_drawdowns)

# Compare portfolios
print("PORTFOLIO COMPARISON")
print("="*60)
print(f"{'Metric':<25} {'Current':>15} {'Equal Weight':>15}")
print("-"*60)
print(f"{'Ann. Return':<25} {port_ann_return:>14.2%} {eq_ann_return:>14.2%}")
print(f"{'Ann. Volatility':<25} {port_ann_vol:>14.2%} {eq_ann_vol:>14.2%}")
print(f"{'Sharpe Ratio':<25} {port_sharpe:>15.4f} {eq_sharpe:>15.4f}")
print(f"{'Max Drawdown':<25} {max_drawdown:>14.2%} {eq_max_drawdown:>14.2%}")

# Identify over/underweight positions
print("\nWEIGHT DIFFERENCES (Current vs Equal)")
print("-"*60)
weight_diff = weights - eq_weights
overweight_mask = weight_diff > 0
underweight_mask = weight_diff < 0

print("Overweight Positions:")
overweight_stocks = stock_names[overweight_mask]
overweight_diffs = weight_diff[overweight_mask]
for stock, diff in zip(overweight_stocks, overweight_diffs):
    print(f"  {stock}: +{diff:.1%}")

print("\nUnderweight Positions:")
underweight_stocks = stock_names[underweight_mask]
underweight_diffs = weight_diff[underweight_mask]
for stock, diff in zip(underweight_stocks, underweight_diffs):
    print(f"  {stock}: {diff:.1%}")

# Interpretation
print("\nINTERPRETATION")
print("-"*60)
print("The equal-weight portfolio achieves a slightly different risk-return profile. Depending on the exact numbers, one may offer a better Sharpe ratio or lower maximum drawdown, indicating whether the active weights in the current portfolio are truly adding value over a naive allocation.")
```
